```python
from tkinter import *
import time
root = Tk()
root.geometry("300x400")
root.title("wordle game")

word=['s','t','a','r','t']

n=0

def Retry():
    global n
    n=5
    inputtxt.config(state= NORMAL)
    return
    

def Take_input():
    global n
    INPUT = inputtxt.get("1.0", "end-1c")
    INPUT=INPUT.lower()
    print(INPUT)
    
    if len(INPUT)!=5:
        Output.delete("1.0","end")
        Output.insert(END,"Word length should be 5")
        return
        
    if n>5:
        Output.delete("1.0","end")
        Output.insert(END,"Sorry Your all attempts are OVER Click on Retry")
        inputtxt.config(state= "disabled") 
        
    INPUT=list(INPUT)
    print(INPUT)
    n+=1
    if(INPUT ==word):
        Output.delete("1.0","end")
        Output.insert(END,"Great it's the same WORD")
        inputtxt.config(state= "disabled")
       
    else:
        print("its not correct")
        Output.delete("1.0","end")
        for j in range(5):
            if INPUT[j]!=word[j]:
                INPUT[j]='_'
                Output.insert(END,INPUT[j])
            else:
                Output.insert(END,INPUT[j])
                Output.tag_config("start", foreground="Green")
                s='1.'+str(j)
                e='1.'+str(j+1)
                Output.tag_add("start", s, e)
        Output.insert(END,"\nCharacter in Green are Correct")
     
l = Label(text = "Find the Hidden word of 5 Character")

inputtxt = Text(root, height = 10,width = 25)

Output = Text(root, height = 3,width = 34)

Display = Button(root, height = 2,
                 width = 15,
                 text ="Check the word",
                 command = lambda:Take_input())

Rety= Button(root, height = 2,
                 width = 10,
                 text ="Retry",
                 command = lambda:Retry())
Quit= Button(root, height = 2,
                 width = 10,
                 text ="Quit",
                 command =root.destroy)
 
l.pack()
inputtxt.pack()
Display.pack()
Output.pack()
Rety.pack(side=LEFT)
Quit.pack(side=RIGHT)
mainloop()
```
